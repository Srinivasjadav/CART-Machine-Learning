```python
import csv
import random
import pandas as pd
import numpy as np
```

```python
seic_benefit_list = []

health_list = []
health_benefit_list = []

house_list = []
house_benefit_list =[]

nutrition_list = []
nutrition_benefit_list = []

livelihood_list = []
livelihood_benefit_list = []

education_list= []
education_benefit_list = []

scheme1_list = []
scheme1_benefit_list = []

scheme2_list =[]
scheme2_benefit_list = []

scheme_choice = ["Eligible" , "Non-Eligible"]

for i in range(0,10000):
    seic_benefit = random.randint(0,100000)
    seic_benefit_list.append(seic_benefit)
    
    health = random.randint(0,100)
    health_list.append(health)
    
    health_benefit = random.randint(0,80000)
    health_benefit_list.append(health_benefit)
    
    house = random.randint(0,100)
    house_list.append(house)
    
    house_benefit = random.randint(0,80000)
    house_benefit_list.append(house_benefit)
    
    
    nutrition = random.randint(0,100)
    nutrition_list.append(nutrition)
    
    
    nutrition_benefit = random.randint(0,80000)
    nutrition_benefit_list.append(nutrition_benefit)
    
    
    livelihood = random.randint(0,100)
    livelihood_list.append(livelihood)
    
    livelihood_benefit = random.randint(0,100000)
    livelihood_benefit_list.append(livelihood_benefit)

    
    education = random.randint(0,100)
    education_list.append(education)
    
    education_benefit= random.randint(0,80000)
    education_benefit_list.append(education_benefit)
    
    
    scheme1 = random.randint(0,100)
    scheme1_list.append(scheme1)
    
    scheme1_benefit = random.choice(scheme_choice)
    scheme1_benefit_list.append(scheme1_benefit)
    
    scheme2=  random.randint(0,100)
    scheme2_list.append(scheme2)

    scheme2_benefit= random.choice(scheme_choice)
    scheme2_benefit_list.append(scheme2_benefit)
    
    
dict = { "seic_benefit" : seic_benefit_list,
        "health" :health_list ,
        "health_benefit" : health_benefit_list,
       "house": house_list,
       "house_benefit" : house_benefit_list,
       "nutrition" : nutrition_list,
        "nutrition_benefit" :nutrition_benefit_list,
        "livelihood": livelihood_list,
        "livelihood_benefit" : livelihood_benefit_list,
        "education" : education_list,
        "education_benefit" : education_benefit_list,
        "scheme1" : scheme1_list,
        "scheme1_benefit" : scheme1_benefit_list,
        "scheme2" : scheme2_list,
        "scheme2_benefit" : scheme2_benefit_list
       }

df = pd.DataFrame(dict)

df.shape
```

```python
seic = list(range(10000))
import math 

for i in range(10000):
    seic[i] = math.floor((health_list[i] + house_list[i] +
                nutrition_list[i] + livelihood_list[i] +
                education_list[i] )/5)
    
len(seic)

df.head()
```

```python
df.describe()
```

```python
df["seic"] = seic

len(df)
```

```python
df.head()
```

```python
seic_error = list(range(10000))

for i in range(10000):
    if ((df.seic[i] >=70) & (df.seic_benefit[i] >=50000)):
        seic_error[i] = "No Error"
        
    elif ((df.seic[i] >=70) & (df.seic_benefit[i] <=20000)):
        seic_error[i] = "EOE"
        
    elif ((df.seic[i] <=30) & (df.seic_benefit[i] >=50000)):
        seic_error[i] = "EOI"
        
    elif ((df.seic[i] <=30) & (df.seic_benefit[i] <=20000)):
        seic_error[i] = "No Error"
        
    else:
        seic_error[i] = "No Error"
    


df["SEIC ERROR"] = seic_error

len(df)
```

```python
health_error = list(range(10000))

for i in range(10000):
    if ((df.health[i] >=75) & (df.health_benefit[i] >=40000)):
        health_error[i] = "No Error"
        
    elif ((df.health[i] >=75) & (df.health_benefit[i] <=10000)):
        health_error[i] = "EOE"
        
    elif ((df.health[i] <=25) & (df.health_benefit[i] >=40000)):
        health_error[i] = "EOI"
        
    elif ((df.health[i] <=25) & (df.health_benefit[i] <=10000)):
        health_error[i] = "No Error"
        
    else:
        health_error[i] = "No Error"
        
df["Health ERROR"] = health_error
len(df)
```
